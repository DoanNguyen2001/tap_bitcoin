```python
def last_unique_letter(s):
    unique_letters = ''.join(dict.fromkeys(s[::-1]))  # Reverse string and remove duplicates
    return unique_letters[0] if unique_letters else None
```

```python
# Function to convert 'K' and 'M' to numeric values
def convert_to_numeric(val):
    if isinstance(val, str):
        if 'K' in val:
            return float(val.replace('K', '')) * 1e3
        elif 'M' in val:
            return float(val.replace('M', '')) * 1e6
    return float(val)
# Apply the conversion to the 'vol.' column
df['vol.'] = df['vol.'].apply(convert_to_numeric)
```

```python
# Define a function to preprocess comments by converting them to lowercase and removing punctuation
import string

def preprocessing(comment):
    # Check if the input is a string and not null
    if isinstance(comment, str):
        # Convert comment to lowercase
        comment_lower = comment.lower()
        # Remove punctuation from the lowercase comment
        comment_no_punc = comment_lower.translate(str.maketrans("", "", string.punctuation))
        # Return preprocessed comment
        return comment_no_punc
    elif pd.isnull(comment):
        # If the value is null (None or NaN), return an empty string or handle as needed
        return ""
    else:
        # If it's not a string but also not null, return the value unchanged
        return comment
```

```python
# import stopwords
from nltk.corpus import stopwords
stopwords_list = stopwords.words('english')
```

```python
# Importing necessary module
import nltk
from nltk.corpus import wordnet

# Defining function to get the WordNet POS tag for lemmatization
def get_wordnet_pos(word):
    # Getting the POS tag using NLTK's pos_tag function
    tag = nltk.pos_tag([word])[0][1][0].upper()
    # Creating a dictionary to map NLTK's POS tags to WordNet's POS tags
    tag_dict = {"J": wordnet.ADJ,  # Adjective
                "N": wordnet.NOUN,  # Noun
                "V": wordnet.VERB,  # Verb
                "R": wordnet.ADV}   # Adverb
    
    # Returning the corresponding WordNet POS tag, defaulting to NOUN if not found
    return tag_dict.get(tag, wordnet.NOUN)

# Creating a WordNet Lemmatizer object
lemmatizer = nltk.stem.WordNetLemmatizer()

# Defining function to lemmatize a sentence
def get_lemmatize(sentence):        
    # Lemmatizing each word in the sentence using the appropriate WordNet POS tag
    return " ".join([lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in sentence])
```
